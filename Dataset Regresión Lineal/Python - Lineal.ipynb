<h2>Dataset Regresión Lineal</h2>

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
import seaborn as ans
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df
```

<h2>Grafica de dispersion</h2>

```python
df.plot.scatter(x="precio", y="metro")
plt.show()
```

```python
df.columns
df.info
df.describe
```

```python
regresion = linear_model.LinearRegression()
```

```python
years = df["metro"].values.reshape((-1,1))
```

```python
modelo = regresion.fit(years, df["precio"])
print("Interseccion (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
```

```python
entrada = [[1],[15],[25],[30]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
df.plot.scatter(x="metro", y="precio", label='Datos Originales')
plt.scatter(entrada, predicciones, color='red')
plt.plot(entrada, predicciones, color='black', label='Linea de regresión logistica')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
