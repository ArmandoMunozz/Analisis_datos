<h2>Dataset Regresión Logística</h2>

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
import seaborn as ans
import pandas as pd
```

```python
df = pd.read_csv("framingham.csv")
df
```

<h2>Grafica de dispersion</h2>

```python
df.plot.scatter(x="age", y="currentSmoker")
plt.show()
```

```python
df.columns
df.info
df.describe
```

```python
regresion = linear_model.LinearRegression()
```

```python
years = df["age"].values.reshape((-1,1))
```

```python
modelo = regresion.fit(years, df["currentSmoker"])
print("Interseccion (b)", modelo.intercept_)
print("Pendiente (m)", modelo.coef_)
```

```python
entrada = [[32],[40],[52],[70]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
df.plot.scatter(x="age", y="currentSmoker", label='Datos Originales')
plt.scatter(entrada, predicciones, color='red')
plt.plot(entrada, predicciones, color='black', label='Linea de regresión logistica')
plt.xlabel('age')
plt.ylabel('currentSmoker')
plt.legend()
plt.show()
```
